```python
# Indicator Analysis

# The first step is to import the libraries required to conduct a detailed analysis.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# I will now load the .xlsx file and use .head to take a preliminary look at the data.
df = pd.read_excel('Analista__De_Consumo.xlsx')
df.head()
```

```python
# Here I load the sheets from the file.
num_anual_df = pd.read_excel('Analista__De_Consumo.xlsx', sheet_name='NumAnual')
num_mensual_df = pd.read_excel('Analista__De_Consumo.xlsx', sheet_name='NumMensual')
```

```python
# I will begin by working with the consumption indicators.

# I will create a chart to show water consumption by Business Unit, 
# allowing us to identify which units consume the most and which consume the least.

# I filter the data by cubic meters (m³).
consumo_agua_negocio = num_anual_df[num_anual_df['AbreviaturaUnidadMedida'] == 'm3'].groupby('NombreTema')['RtaNumericaEvento_t'].sum().reset_index()

# Bar chart.
plt.figure(figsize=(6, 4))
plt.bar(consumo_agua_negocio['NombreTema'], consumo_agua_negocio['RtaNumericaEvento_t'], color='skyblue')
plt.title('Total Water Consumption by Business Unit (m³)')
plt.ylabel('Total Consumption (m³)')
plt.xlabel('Business')
plt.xticks(rotation=45)
plt.show()

```

```python
# I will do the same to analyze energy consumption.

# First, I filter the data by kWh.
consumo_energia_negocio = num_anual_df[num_anual_df['AbreviaturaUnidadMedida'] == 'KWh'].groupby('NombreTema')['RtaNumericaEvento_t'].sum().reset_index()

# Bar chart.
plt.figure(figsize=(6, 4))
plt.bar(consumo_energia_negocio['NombreTema'], consumo_energia_negocio['RtaNumericaEvento_t'], color='yellow')
plt.title('Total Water Consumption by Business Unit (KWh)')
plt.ylabel('Total Consumption (KWh)')
plt.xlabel('Business')
plt.xticks(rotation=45)
plt.show()
```

```python
# I will now create a chart to show which resource has the greatest impact
# on the consumption indicator.

# I group the data by resource type and sum the consumption.
consumo_por_recurso = num_anual_df.groupby('NombreTabla')['RtaNumericaEvento_t'].sum().reset_index()

# Bar chart.
plt.figure(figsize=(6, 4))
plt.bar(consumo_por_recurso['NombreTabla'], consumo_por_recurso['RtaNumericaEvento_t'], color='salmon')
plt.title('# Total Consumption by Resource Type')
plt.ylabel('Total Consumption (RtaNumericaEvento_t)')
plt.xlabel('Resource Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# I will now create a line chart to show water consumption by Business Unit
# between the years 2022 and 2023.

# First, I filter the data by cubic meters (m³).
consumo_agua = num_anual_df[num_anual_df['AbreviaturaUnidadMedida'] == 'm3']

# I group the data by year and category, summing the consumption.

consumo_agua_por_negocio_en_años = consumo_agua.groupby(['Evento_t', 'NombreTema'])['RtaNumericaEvento_t'].sum().unstack()

# I create the line chart.
consumo_agua_por_negocio_en_años.plot(kind='line', figsize=(6, 4))

# With this code, I design the chart.
plt.title('Water Consumption Trend by Business Unit in 2022 and 2023 (m³)')
plt.ylabel('Total Consumption (m³)')
plt.xlabel('Year')
plt.legend(title='Business')
plt.show()
```

```python
# I will do the same with the energy indicator.


# First, we filter the data by kWh.
consumo_energia = num_anual_df[num_anual_df['AbreviaturaUnidadMedida'] == 'KWh']

# I group the data by year and category, summing the consumption.
consumo_energia_por_negocio_en_años = consumo_energia.groupby(['Evento_t', 'NombreTema'])['RtaNumericaEvento_t'].sum().unstack()

# I create the line chart.
consumo_energia_por_negocio_en_años.plot(kind='line', figsize=(6, 4))

# With this code, I design the chart.
plt.title('Energy Consumption Trend by Business Unit in 2022 and 2023 (KWh)')
plt.ylabel('Total Consumption (KWh)')
plt.xlabel('Year')
plt.legend(title='Business')
plt.show()
```

```python
# To find the variation in consumption from one year to another, I first create
# a pivot table that shows the consumption by plant.

# First, I filter the data by "Water Consumption" only.

variazion_df = pd.pivot_table(num_anual_df[num_anual_df['NombreTabla'] == 'Consumo de agua'], values= ['RtaNumericaEvento_t'], index= ['NombreFormulario'], columns=['Evento_t'], aggfunc= np.sum)

# And I add the variation to the pivot table.
variazion_df['Variación (%)'] = ((variazion_df[('RtaNumericaEvento_t', 2023)] - variazion_df[('RtaNumericaEvento_t', 2022)]) / variazion_df[('RtaNumericaEvento_t', 2022)]) * 100
variazion_df
```

```python
# I do the same for the category "Energy Consumption from Non-Renewable Sources."
variazion_df = pd.pivot_table(num_anual_df[num_anual_df['NombreTabla'] == 'Consumo de energía de fuentes no renovables'], values= ['RtaNumericaEvento_t'], index= ['NombreFormulario'], columns=['Evento_t'], aggfunc= np.sum)
variazion_df['Variación (%)'] = ((variazion_df[('RtaNumericaEvento_t', 2023)] - variazion_df[('RtaNumericaEvento_t', 2022)]) / variazion_df[('RtaNumericaEvento_t', 2022)]) * 100
variazion_df
```

```python
# Now, I will analyze the indicators from the 'NumMensual' sheet.

# I start by looking at the number of hours and days worked, as well as the number of workers.
# I will create a pivot table to show the totals by month for each Business Unit.
indicadores_social = pd.pivot_table(num_mensual_df, values= ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'], index= ['NombreTema','Nombre fila'], aggfunc= np.sum)

# Dictionary for converting months from Spanish to English.
meses_dict = {
    'Enero': 'January',
    'Febrero': 'February',
    'Marzo': 'March',
    'Abril': 'April',
    'Mayo': 'May',
    'Junio': 'June',
    'Julio': 'July',
    'Agosto': 'August',
    'Septiembre': 'September',
    'Octubre': 'October',
    'Noviembre': 'November',
    'Diciembre': 'December'
}

# I convert the month names from Spanish to English in the pivot table columns.
indicadores_social.columns = indicadores_social.columns.map(meses_dict)

# I reorder the columns according to the month order in English.
orden_meses = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
indicadores_social = indicadores_social[orden_meses]
indicadores_social


```

```python
# I will plot the data presented in the pivot table for better interpretation.
# First, I will do this with the "Days Worked" indicator.

# I need to filter the values by the mentioned indicator only.
dias_trabajados_df = indicadores_social.loc[(slice(None), 'Días trabajados'), :]

# To achieve an optimal result, I need to convert 'NombreTema' into a column, 
# as it appears as an index in the pivot table, and I wouldn't be able to get the desired result.
dias_trabajados_df_reset = dias_trabajados_df.reset_index(level=1, drop=True)

# Once I have converted 'NombreTabla' into a column, I now transpose the pivot table.
dias_trabajados_df_transposed = dias_trabajados_df_reset.T

# And now I plot the graph.

plt.figure(figsize=(6, 4))
plt.plot(dias_trabajados_df_transposed, marker='o')
plt.legend(dias_trabajados_df_transposed.columns, title='Business')
plt.title('Days Worked by Month by Business Unit in 2023')
plt.ylabel('Days Worked')
plt.xticks(ticks=range(12), labels=dias_trabajados_df.columns, rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# I apply the same code as before to visualize the data now by Hours Worked.
horas_trabajadas_df = indicadores_social.loc[(slice(None), 'Horas trabajadas'), :]
horas_trabajadas_df_reset = horas_trabajadas_df.reset_index(level=1, drop=True)
horas_trabajadas_df_transposed = horas_trabajadas_df_reset.T
plt.figure(figsize=(6, 4))
plt.plot(horas_trabajadas_df_transposed, marker='o')
plt.legend(horas_trabajadas_df_transposed.columns, title='Business')
plt.title('Hours Worked by Month by Business Unit in 2023')
plt.ylabel('Hours Worked')
plt.xticks(ticks=range(12), labels=horas_trabajadas_df.columns, rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Finally, we apply the same code to plot the values by the number of workers
# for each business unit by month.
numero_trabajadores_df = indicadores_social.loc[(slice(None), 'Número total de trabajadores'), :]
numero_trabajadores_df_reset = numero_trabajadores_df.reset_index(level=1, drop=True)
numero_trabajadores_df_transposed = numero_trabajadores_df_reset.T
plt.figure(figsize=(6, 4))
plt.plot(numero_trabajadores_df_transposed, marker='o')
plt.legend(numero_trabajadores_df_transposed.columns, title='Business')
plt.title('Number of Workers by Month by Business Unit in 2023')
plt.ylabel('Number of Workers')
plt.xticks(ticks=range(12), labels=numero_trabajadores_df.columns, rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
